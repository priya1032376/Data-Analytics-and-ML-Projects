```python
# Importing all the necessary libraries for EDA and visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
import scipy.stats as stats
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# importinga nd reading data 
my_df = pd.read_csv('C:/Users/mishrap6/OneDrive - Zimmer Biomet/Downloads/DS1_C5_S5_Hotel Dataset (5).csv')
my_df.head()

#imputing missing values :
my_df['agent'].fillna(my_df['agent'].mean(), inplace = True) 
my_df['company'].fillna(my_df['company'].mean(), inplace = True)

print(my_df.isnull().sum())
```

```python
# Get the value counts of total_of_special_requests
special_requests_count = my_df['total_of_special_requests'].value_counts().sort_index()

# Create a bar graph for special requests count
plt.figure(figsize=(8,6))
sns.barplot(x=special_requests_count.index, y=special_requests_count.values, palette='viridis')
plt.title('Count of Special Requests', fontsize=14)
plt.xlabel('Total Special Requests', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# special requests based on room type using a pivot table or crosstab.

# Create a pivot table for reserved_room_type and total_of_special_requests
special_requests_by_room_type = pd.crosstab(my_df['reserved_room_type'], my_df['total_of_special_requests'])

# Plot a bar graph for special requests by room type
special_requests_by_room_type.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set3')
plt.title('Special Requests by Room Type', fontsize=14)
plt.xlabel('Reserved Room Type', fontsize=12)
plt.ylabel('Count of Special Requests', fontsize=12)
plt.legend(title='Total Special Requests')
plt.tight_layout()
plt.show()
```
