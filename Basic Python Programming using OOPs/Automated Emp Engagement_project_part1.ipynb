```python
# Task 1 : write a program to develop a simple console based computer brain teaser, a riddle logic puzzle, text based puzzles, text based puzzle, involving
# word play, logic and letaral thinking games in python. The prgram will randomly choose a riddle question from the list of riddle (que-answer base dictionary)
# provided in the program. It will then prompt the user for the answer and compare it against correct answer.
# If the user's answer matches with correct one it will print "Correct". otherwise it will print "incorrect" and then provide correct answer.
# The program will run infinitely allowing user to solve multiple riddles with the option to play again. Upon exiiting the game it will print user's score
# percentage.

from abc import ABC, abstractmethod
import random




class Riddle_game(ABC):
  def __init__(self, user_name):
    self.user_name = user_name
    self.que_asked = 0
    self.obtained_score = 0
    self.full_score = 0
    self.riddle_qa = {
                      "What has keys but can't open locks?": "piano",
                      "What has a heart that doesn't beat?": "artichoke",
                      "What can travel around the world while staying in a corner?": "stamp",
                      "What comes down but never goes up?": "Rain",
                      "What has an eye but can't see?": "needle"
                  }


  @abstractmethod
  def riddle_match(self, option): pass

  def play_game(self):
    while True:
          question=random.choice(list(self.riddle_qa.keys()))   # getting any random que (key of dic) from the list of key-value pair and passig to variable question
          print(f"Your Question : {question}")
          answer = input("Enter your anwser: ")
          self.full_score+=10

          if answer.lower()==self.riddle_qa[question].lower():
            print(f"Correct!")
            self.obtained_score+=10
          else:
            print(f"Wrong! The correct answer is : {self.riddle_qa[question]}")

          play_again = input("Do you wish to play again?: (Y/N)")

          if play_again.lower()=='n':
           break

    self.percentage = (self.obtained_score/self.full_score)*100
    print(f"Game Over!\nYour score:{self.obtained_score}/{self.full_score}")
    print(f"Percentage: {self.percentage:.2f}%")

```

```python
# Task 2: write a program to develop a simple console based computer brain teaser, an arithmatic logic puzzle, a problem solving game in python.
# the program will randomly generate 2 numerics with range 1 to 20 and and an arithmetic operator (+,-, /, *). The program will runs infinite turns with the
# option play again.

# The program should handle exceptions gracefully,  such as "invalid input" when user attempts to solve the puzzle. and program should print the percentage when
# user leave the game.

from abc import ABC, abstractmethod

class Arithmatic_quiz(ABC):

  def __init__(self, player_name):
    self.player_name = player_name
    self.numbers=list(range(1,21))
    self.operators=["+", "-", "/", "*"]
    self.full_marks = 0
    self.obtained_marks = 0


  @abstractmethod
  def check_answer(self): pass


  def play_game(self):
    while True:

      self.num1= str(random.choice(self.numbers))
      self.operator1= random.choice(self.operators)
      self.num2 = str(random.choice(self.numbers))
      self.question = self.num1 + self.operator1 + self.num2
      print(f"Question : {self.question}")
      self.answer  = float(eval(self.question))
      user_answer = float(input("Enter your answer : "))
      if self.answer == user_answer:
        print("Correct!")
        self.full_marks+=10
        self.obtained_marks+=10

      else:
        print(f"Wrong! the correct answer is : {self.answer}")
        self.full_marks+=10
        self.obtained_marks+=0

      play_again = input("Do You want to play again? (Yes/No)").lower()
      if play_again != "yes":
        break


    self.percentage = (self.obtained_marks/self.full_marks)*100
    print(f"Game Over!\nYour score:{self.obtained_marks}/{self.full_marks}")
    print(f"Percentage: {self.percentage:.2f}%")





```

```python
class Playnow(Arithmatic_quiz):

  def check_answer(self): pass

if __name__ == '__main__':
  player1 = Playnow("Payer ABC")

  player1.play_game()

```

```python
# Task 3 : write a program to develop a general knowledge quiz game.create 4 lists to store GK ques of 4 category - science, math, geography and other.
# Each items in thsese list is adictionary with "que" and "answer" keys and provided in the sides, Ensure to use proper exception handling to handle invalid
# user input.

# the program should display ques based on category bseed on selected category by the user. after user submit the answer ot should indiate weather it is correct
#if the answer is incorrect it should display the correct answer.


# List of Q/A :
# Sicence category :

# que - which gas do plant absorb during photosynthesis : Ans - Carbon dioxide
# Que - What is the name of WWW - Ans - World wide web
# Ques - How many bones are there in adult human body?  Ans  - 206

# MAth catecory
# Que- what is the only even prime number - ans - 2
# ques - what is the sum of first 5 natural numbers? 15
# ques - what is the name of mathematical symbol π ? Ans - Pi

# geography category ?
# Which is the tallest mountain on earth ? Mount everest
# what is the Capital of france? Paris
# Which planet is known as red planet? Mars


# others :
# Which country win the most recent FIFA world cup (As of 2020) ? Argentina
# Who invent the light bulb ? Thomas Edison
# In what year did India decalre itself a democratic and Republic state with adoption of constitution? 1950


from abc import ABC, abstractmethod
import random

class GK(ABC):
    def __init__(self, player_name):
        self.player_name = player_name
        self.full_score = 0
        self.obtained_score = 0
        self.science = {
            "which gas do plants absorb during photosynthesis?": "Carbon dioxide",
            "What is the name of WWW?": "World wide web",
            "How many bones are there in the adult human body?": "206"
        }

        self.maths = {
            "What is the only even prime number?": "2",
            "What is the sum of the first 5 natural numbers?": "15",
            "What is the name of the mathematical symbol π?": "Pi"
        }

        self.geography = {
            "Which is the tallest mountain on earth?": "Mount Everest",
            "What is the capital of France?": "Paris",
            "Which planet is known as the red planet?": "Mars"
        }

        self.others = {
            "Which country won the most recent FIFA World Cup (as of 2020)?": "Argentina",
            "Who invented the light bulb?": "Thomas Edison",
            "In what year did India declare itself a democratic and republic state with the adoption of the constitution?": "1950"
        }

    @abstractmethod
    def options(self, option):
        pass

    def play_game(self):
        while True:
            try:
                option = int(input("Welcome to the General Knowledge Game.\n"
                                   "Select a category:\n"
                                   "Press 1 for Science\n"
                                   "Press 2 for Maths\n"
                                   "Press 3 for Geography\n"
                                   "Press 4 for Others\n"
                                   ": "))
                if option not in [1, 2, 3, 4]:
                    print("Invalid Input, please enter a number between 1 and 4.")
                    continue
                break
            except ValueError:
                print("Invalid Input, please enter a number between 1 and 4.")

        # Science category
        if option == 1:
            while True:
                question_science = random.choice(list(self.science.keys()))
                print(f"Your Question: {question_science}")
                answer_science = input("Enter your answer: ")

                if answer_science.lower() == self.science[question_science].lower():
                    print("Correct!")
                    self.obtained_score += 10
                    self.full_score += 10
                else:
                    self.full_score += 10
                    print(f"Wrong! The correct answer is: {self.science[question_science]}")

                play_again = input("Do you wish to continue playing in the same category? (Y/N): ")
                if play_again.lower() != "y":
                    break

            print(f"Your Science Score is: {self.obtained_score}/{self.full_score}")

        # Maths category
        elif option == 2:
            while True:
                question_maths = random.choice(list(self.maths.keys()))
                print(f"Your Question: {question_maths}")
                answer_maths = input("Enter your answer: ")

                if answer_maths.lower() == self.maths[question_maths].lower():
                    print("Correct!")
                    self.obtained_score += 10
                    self.full_score += 10
                else:
                    self.full_score += 10
                    print(f"Wrong! The correct answer is: {self.maths[question_maths]}")

                play_again = input("Do you wish to continue playing in the same category? (Y/N): ")
                if play_again.lower() != "y":
                    break

            print(f"Your Maths Score is: {self.obtained_score}/{self.full_score}")

        # Geography category
        elif option == 3:
            while True:
                question_geography = random.choice(list(self.geography.keys()))
                print(f"Your Question: {question_geography}")
                answer_geography = input("Enter your answer: ")

                if answer_geography.lower() == self.geography[question_geography].lower():
                    print("Correct!")
                    self.obtained_score += 10
                    self.full_score += 10
                else:
                    self.full_score += 10
                    print(f"Wrong! The correct answer is: {self.geography[question_geography]}")

                play_again = input("Do you wish to continue playing in the same category? (Y/N): ")
                if play_again.lower() != "y":
                    break

            print(f"Your Geography Score is: {self.obtained_score}/{self.full_score}")

        # Others category
        elif option == 4:
            while True:
                question_others = random.choice(list(self.others.keys()))
                print(f"Your Question: {question_others}")
                answer_others = input("Enter your answer: ")

                if answer_others.lower() == self.others[question_others].lower():
                    print("Correct!")
                    self.obtained_score += 10
                    self.full_score += 10
                else:
                    self.full_score += 10
                    print(f"Wrong! The correct answer is: {self.others[question_others]}")

                play_again = input("Do you wish to continue playing in the same category? (Y/N): ")
                if play_again.lower() != "y":
                    break

            print(f"Your Others Score is: {self.obtained_score}/{self.full_score}")

        while True:
            choose = input("Enter 'Yes' if you wish to play again in a different category. \nEnter 'Exit' if you wish to exit the game: ")
            if choose.lower() == "exit":
                break
            elif choose.lower() == "yes":
                self.play_game()
                break
            else:
                print("Invalid input, please enter 'Yes' or 'Exit'.")

        if self.full_score > 0:
            self.percentage = (self.obtained_score / self.full_score) * 100
        else:
            self.percentage = 0
        print(f"Game Over!\nYour score: {self.obtained_score}/{self.full_score}")
        print(f"Percentage: {self.percentage:.2f}%")
        return


class Playnow(GK):
    def options(self, option):
        pass


if __name__ == '__main__':
    player1 = Playnow("Player ABC")
    player1.play_game()
```

```python
# Task 5 : Stome paper scissor for 1 user :

import random

def get_computer_choice():
    """Function to generate computer's random choice."""
    return random.choice(["rock", "paper", "scissors"])

def get_user_choice():
    """Function to get the user's choice."""
    while True:
        user_input = input("Enter Rock/Paper/Scissors: ").lower()
        if user_input in ["rock", "paper", "scissors"]:
            return user_input
        else:
            print("Invalid choice! Please try again.")

def determine_winner(computer_choice, user_choice):
    """Function to determine the winner based on the choices."""
    if computer_choice == user_choice:
        return "It's a tie!"

    # Define the winning conditions
    if (computer_choice == "rock" and user_choice == "scissors") or \
       (computer_choice == "paper" and user_choice == "rock") or \
       (computer_choice == "scissors" and user_choice == "paper"):
        return "Computer wins!"
    else:
        return "You win!"

def play_game():
    """Function to play the Rock, Paper, Scissors game."""
    while True:
        print("Welcome to Rock, Paper, Scissors!")
        computer_choice = get_computer_choice()
        user_choice = get_user_choice()

        print(f"Computer chose: {computer_choice}")
        print(f"You chose: {user_choice}")

        # Determine the winner
        result = determine_winner(computer_choice, user_choice)
        print(result)

        # Ask if the user wants to play again
        play_again = input("Do you want to play again? (y/n): ").lower()
        if play_again != 'y':
            print("Thanks for playing!")
            break

if __name__ == '__main__':
    play_game()
```
