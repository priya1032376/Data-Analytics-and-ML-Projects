```python
# Task 2

import random
from abc import ABC, abstractmethod

# Base Game Class
class Game(ABC):
    """Abstract base class for games."""

    @abstractmethod
    def play_game(self):
        pass

# Riddle Game Class
class RiddleGame(Game):
    def __init__(self, user_name):
        self.user_name = user_name
        self.que_asked = 0
        self.obtained_score = 0
        self.full_score = 0
        self.riddle_qa = {
            "What has keys but can't open locks?": "piano",
            "What has a heart that doesn't beat?": "artichoke",
            "What can travel around the world while staying in a corner?": "stamp",
            "What comes down but never goes up?": "Rain",
            "What has an eye but can't see?": "needle"
        }

    def play_game(self):
        while True:
            question = random.choice(list(self.riddle_qa.keys()))
            print(f"Your Question: {question}")
            answer = input("Enter your answer: ")
            self.full_score += 10

            if answer.lower() == self.riddle_qa[question].lower():
                print("Correct!")
                self.obtained_score += 10
            else:
                print(f"Wrong! The correct answer is: {self.riddle_qa[question]}")

            play_again = input("Do you wish to play again? (Y/N): ").lower()
            if play_again == 'n':
                break

        percentage = (self.obtained_score / self.full_score) * 100
        print(f"Game Over!\nYour score: {self.obtained_score}/{self.full_score}")
        print(f"Percentage: {percentage:.2f}%")

# Arithmetic Quiz Game Class
class ArithmeticQuiz(Game):
    def __init__(self, player_name):
        self.player_name = player_name
        self.numbers = list(range(1, 21))
        self.operators = ["+", "-", "/", "*"]
        self.full_marks = 0
        self.obtained_marks = 0

    def play_game(self):
        while True:
            num1 = str(random.choice(self.numbers))
            operator1 = random.choice(self.operators)
            num2 = str(random.choice(self.numbers))
            question = num1 + operator1 + num2
            print(f"Question: {question}")
            answer = float(eval(question))
            user_answer = float(input("Enter your answer: "))

            if answer == user_answer:
                print("Correct!")
                self.full_marks += 10
                self.obtained_marks += 10
            else:
                print(f"Wrong! The correct answer is: {answer}")
                self.full_marks += 10

            play_again = input("Do you want to play again? (Yes/No): ").lower()
            if play_again != "yes":
                break

        percentage = (self.obtained_marks / self.full_marks) * 100
        print(f"Game Over!\nYour score: {self.obtained_marks}/{self.full_marks}")
        print(f"Percentage: {percentage:.2f}%")

# General Knowledge Game Class
class GeneralKnowledge(Game):
    def __init__(self, player_name):
        self.player_name = player_name
        self.full_score = 0
        self.obtained_score = 0
        self.categories = {
            1: {
                "name": "Science",
                "questions": {
                    "Which gas do plants absorb during photosynthesis?": "Carbon dioxide",
                    "What is the name of WWW?": "World Wide Web",
                    "How many bones are there in the adult human body?": "206"
                }
            },
            2: {
                "name": "Maths",
                "questions": {
                    "What is the only even prime number?": "2",
                    "What is the sum of the first 5 natural numbers?": "15",
                    "What is the name of the mathematical symbol Ï€?": "Pi"
                }
            },
            3: {
                "name": "Geography",
                "questions": {
                    "Which is the tallest mountain on earth?": "Mount Everest",
                    "What is the capital of France?": "Paris",
                    "Which planet is known as the red planet?": "Mars"
                }
            },
            4: {
                "name": "Others",
                "questions": {
                    "Which country won the most recent FIFA World Cup (as of 2020)?": "Argentina",
                    "Who invented the light bulb?": "Thomas Edison",
                    "In what year did India declare itself a democratic and republic state with the adoption of the constitution?": "1950"
                }
            }
        }

    def play_game(self):
        while True:
            try:
                option = int(input("Welcome to the General Knowledge Game.\n"
                                   "Select a category:\n"
                                   "1. Science\n"
                                   "2. Maths\n"
                                   "3. Geography\n"
                                   "4. Others\n"
                                   "Enter your choice: "))
                if option not in [1, 2, 3, 4]:
                    print("Invalid Input, please enter a number between 1 and 4.")
                    continue
            except ValueError:
                print("Invalid Input, please enter a number between 1 and 4.")
                continue

            category = self.categories[option]
            while True:
                question = random.choice(list(category["questions"].keys()))
                print(f"Your Question: {question}")
                answer = input("Enter your answer: ")

                if answer.lower() == category["questions"][question].lower():
                    print("Correct!")
                    self.obtained_score += 10
                else:
                    print(f"Wrong! The correct answer is: {category['questions'][question]}")
                self.full_score += 10

                play_again = input(f"Do you wish to continue playing in the {category['name']} category? (Y/N): ").lower()
                if play_again != "y":
                    break

            print(f"Your {category['name']} Score: {self.obtained_score}/{self.full_score}")

            choose = input("Enter 'Yes' to play another category or 'Exit' to quit: ").lower()
            if choose == "exit":
                break
            elif choose == "yes":
                continue
            else:
                print("Invalid input, returning to main menu.")
                break

        percentage = (self.obtained_score / self.full_score) * 100 if self.full_score > 0 else 0
        print(f"Game Over!\nYour score: {self.obtained_score}/{self.full_score}")
        print(f"Percentage: {percentage:.2f}%")

# Rock Paper Scissors Game Class
class RockPaperScissors(Game):
    def get_choice(self, player_name):
        while True:
            choice = input(f"{player_name}, enter Rock/Paper/Scissors: ").lower()
            if choice in ["rock", "paper", "scissors"]:
                return choice
            else:
                print("Invalid choice! Please try again.")

    def determine_winner(self, choice1, choice2):
        if choice1 == choice2:
            return "It's a tie!"
        if (choice1 == "rock" and choice2 == "scissors") or \
           (choice1 == "paper" and choice2 == "rock") or \
           (choice1 == "scissors" and choice2 == "paper"):
            return "Player 1 wins!"
        else:
            return "Player 2 wins!"

    def play_game(self):
        print("Welcome to Rock, Paper, Scissors!")
        player1_name = input("Enter the name of Player 1: ")
        player2_name = input("Enter the name of Player 2: ")

        while True:
            player1_choice = self.get_choice(player1_name)
            player2_choice = self.get_choice(player2_name)

            print(f"{player1_name} chose: {player1_choice}")
            print(f"{player2_name} chose: {player2_choice}")

            result = self.determine_winner(player1_choice, player2_choice)
            print(result)

            play_again = input("Do you want to play again? (y/n): ").lower()
            if play_again != 'y':
                print("Thanks for playing!")
                break

# Word Guessing Game Class
class WordGame(Game):
    def __init__(self, player_name):
        self.player_name = player_name
        self.words = ["Python", "Excel", "PLSQL", "Tableau", "Programming",
                      "Data Science", "Data Analytics", "Business Analytics"]
        self.attempts = 3
        self.computer_choice = None

    def word_choice(self):
        self.computer_choice = random.choice(self.words)
        return self.computer_choice

    def give_hint(self):
        return f"The word starts with '{self.computer_choice[0]}' and has {len(self.computer_choice)} letters."

    def play_game(self):
        self.word_choice()
        print(f"Hint: {self.give_hint()}")

        attempts_left = self.attempts
        while attempts_left > 0:
            guess = input("Enter your guess: ")
            if guess.lower() == self.computer_choice.lower():
                print("Congratulations! You guessed the word!")
                break
            else:
                attempts_left -= 1
                print(f"Wrong guess! You have {attempts_left} attempts left.")

        if attempts_left == 0:
            print(f"Sorry, you are out of attempts. The correct word was: {self.computer_choice}")

# Main Menu Function
def main_menu():
    print("Welcome to the Ultimate Game Collection!")
    player_name = input("Enter your name: ")

    games = {
        1: RiddleGame(player_name),
        2: ArithmeticQuiz(player_name),
        3: GeneralKnowledge(player_name),
        4: RockPaperScissors(),
        5: WordGame(player_name)
    }

    while True:
        print("\nChoose a game to play:")
        print("1. Riddle Game")
        print("2. Arithmetic Quiz")
        print("3. General Knowledge")
        print("4. Rock Paper Scissors")
        print("5. Word Guessing Game")
        print("6. Exit")

        try:
            choice = int(input("Enter the number of the game you want to play: "))
            if choice == 6:
                print("Thanks for playing! Goodbye!")
                break
            elif choice in games:
                games[choice].play_game()
            else:
                print("Invalid choice, please try again.")
        except ValueError:
            print("Invalid input, please enter a number.")

if __name__ == "__main__":
    main_menu()
```
